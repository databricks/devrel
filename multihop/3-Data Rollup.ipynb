### Data Rollup for BI Reporting

```
tbl = databaseName + ".silver_stream"

(spark.readStream
     .format("delta")                               # read delta format
     .table(tbl)                                    # from SILVER table
     .groupBy("unit_num")                           # aggregate data
     .count()                                       # produce counts
     .writeStream                                   # Write the stream
     .format("delta")                               # Use the delta format
     .partitionBy("unit_num")                       # Specify a feature to partition on
     .option("checkpointLocation", goldCheckpointPath)  # Specify where to log metadata
     .option("path", goldOutPath)                   # Specify the output path
     .outputMode("complete")                        # Replace entire table
     .queryName("query_gold")                       # The name of the stream
     .start()                                       # Start the operation
)
```

```
gold_query = 'CREATE TABLE IF NOT EXISTS {}.gold_stream USING delta LOCATION "{}"'.format(databaseName,goldOutPath)
sqlContext.sql(gold_query)
```
