### Data Transformations
* Use 'bronze' table as source of new stream
* Apply business transformations
* Join against reference tables
* Persist to 'silver' tables

```
tbl = databaseName + ".bronze_stream"
df1 = spark.readStream.format("delta").table(tbl)
```

```
display(df1)
```

```
engineMetaDF = spark.sql("SELECT * FROM {}.EngineMetadata".format(databaseName))
display(engineMetaDF)
```

```
from pyspark.sql.functions import *

df2 = df1.filter(df1.unit_num.isNotNull()) \
         .join(engineMetaDF, df1.unit_num == engineMetaDF.engine_id)
```

```
display(df2)
```

```
(df2
  .writeStream                                          # Write the stream
  .format("delta")                                      # Use the delta format
  .option("checkpointLocation", silverCheckpointPath)   # Specify where to log metadata
  .option("path", silverOutPath)                        # Specify the output path
  .outputMode("append")                                 # Append new records to the output path
  .queryName("delta_silver")                            # The name of the stream
  .start()                                              # Start the operation
)
```

```
silver_query = 'CREATE TABLE IF NOT EXISTS {}.silver_stream USING delta LOCATION "{}"'.format(databaseName,silverOutPath)
sqlContext.sql(silver_query)
```
